```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Dataset load ho gaya!")
```

```python
df.head
df.info
df.describe
df.tail
```

```python
print (df.shape)
```

```python
 df.isnull().sum()
```

```python
missing_percent = (df.isnull().sum() /len(df)) * 100
print(missing_percent)
```

```python
len(df)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median()) 
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode())
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df.isnull().sum()
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts(normalize=True) * 100
```

```python
df.groupby('Sex')['Survived'].mean () * 100
```

```python
df.groupby('Pclass')['Survived'].mean () * 100
```

```python
df.groupby('Survived')['Age'].mean () * 100
```

```python

# TITANIC EDA — CONCLUSIONS


# 1. Overall Survival:
#    - 38% log bache, 62% mar gaye

# 2. Gender:
#    - Females mein 74% bachi
#    - Males mein sirf 19% bache
#    - "Women and children first" rule follow hua!

# 3. Passenger Class:
#    - Class 1 (Rich): 63% bache
#    - Class 2 (Middle): 47% bache
#    - Class 3 (Gareeb): 24% bache
#    - Jitna mehenga ticket, utna zyada bachne ka chance!

# 4. Age:
#    - Jo bache unki average age: ~28 saal
#    - Jo nahi bache unki average age: ~30 saal
#    - Age ka zyada farq nahi pada
```
